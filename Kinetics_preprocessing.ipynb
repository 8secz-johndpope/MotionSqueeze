```python
def search(dirname):
    filelist = []
    filenames = os.listdir(dirname)
    for filename in filenames:
        full_filename = os.path.join(dirname, filename)
        ext = os.path.splitext(full_filename)[-1]
        if (ext ==".mp4" or ext == ".mkv" or ext == ".webm"): 
            #print(full_filename)
            filelist.append(full_filename)
    return filelist
```

```python
# Kinetics dataset (validation set)

# for j in range(len(class_names)):
for j in range(1, len(class_names)):    
    dir_name = class_names[j] + "/"#"abseiling/"

    search_default = "/home2/Dataset/Kinetics/compress/val_256/"
    search_name = search_default + dir_name
    #print search_name

    video_file_names = search(search_name)
    file_num =  np.shape(video_file_names)[0]
    print (file_num, video_file_names[0])

    frame_default = "/home/mjkim/Dataset/Kinetics/Kinetics_frames/Kinetics-400_val_frames/"
                                #"/data3/Kinetics_frames_v2/Kinetics-400_val_frames/"
    frame_dir = frame_default + dir_name
    if not os.path.exists(frame_dir):
        os.makedirs(frame_dir)    
    start_time= time.time()
    for i in range(file_num):
#     for i in range(1):
        url = video_file_names[i].split('.m')[0].split('/')[-1] 
        new_dir = frame_dir + url
        if not os.path.exists(new_dir):
            os.makedirs(new_dir)      
        cmd_frames = "ffmpeg -i "+ video_file_names[i]+ " -qscale:v 2 -r 25 " + new_dir+"/img_%05d.jpg"
        call(cmd_frames.split(), shell=False)

    end_time = time.time()    
    print (dir_name, end_time - start_time)
```
